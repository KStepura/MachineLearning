#### 5. [1 балл] Выясните, есть ли студенты с абсолютно одинаковыми предпочтениями по всем курсам.

В данном случае набор курсов задается упорядоченным множеством ('fall_1', 'fall_2', 'fall_3', 'spring_1', 'spring_2', 'spring_3', 'blended'). Элемент blended будет нулевым для 3-го курса.

```python
df_copy = df.copy()
df_copy.loc[df_copy['blended'] == '', 'blended'] = 0
course_sets = df_copy.groupby(['fall_1', 'fall_2', 'fall_3', 'spring_1', 'spring_2', 'spring_3', 'blended']).size().reset_index(name='count')
course_duplicates = course_sets[course_sets['count'] > 1]

#для более красивого и понятного вывода опять использовала библиотеку tabulate
print(tabulate(course_duplicates, headers='keys', tablefmt='pretty', showindex=False))
```

#### 6. [0.5 балла] Найдите курсы по выбору, на которые записывались как студенты 18-го года набора, так и студенты 17-го года.

```python
students_18 = df[df['18_group'] != 0]
students_17 = df[df['17_group'] != 0]

course_columns = ['fall_1', 'fall_2', 'fall_3', 'spring_1', 'spring_2', 'spring_3']

courses_18 = set(students_18[course_columns].stack().unique())
courses_17 = set(students_17[course_columns].stack().unique())

courses_18_and_17 = courses_18.intersection(courses_17)

print("Курсы, на которые записывались как студенты 18-го, так и 17-го годов:")

for i in courses_18_and_17:
    print("-", i)
```

Методом исключения найдите курсы, которые предлагались только студентам 18-го года и только студентам 17-го года.

```python
courses_only_18 = courses_18 - courses_17
courses_only_17 = courses_17 - courses_18

print("Курсы, которые предлагались только студентам 18-го года:")
for i in courses_only_18:
    print("-", i)
print("\nКурсы, которые предлагались только студентам 17-го года:")
for i in courses_only_17:
    print("-", i)
```

### Визуализации и matplotlib

При работе с данными часто неудобно делать какие-то выводы, если смотреть на таблицу и числа в частности, поэтому важно уметь визуализировать данные. Здесь будут описаны ключевые правила оформления графиков для **всех** домашних заданий.

У matplotlib, конечно же, есть [документация](https://matplotlib.org/users/index.html) с большим количеством [примеров](https://matplotlib.org/examples/), но для начала достаточно знать про несколько основных типов графиков:
- plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;
- hist — гистограмма, показывающая распределение некоторой величины;
- scatter — график, показывающий взаимосвязь двух величин;
- bar — столбцовый график, показывающий взаимосвязь количественной величины от категориальной.

В этом задании вы попробуете построить один из них. Не забывайте про базовые принципы построения приличных графиков:
- оси должны быть подписаны, причём не слишком мелко;
- у графика должно быть название;
- если изображено несколько графиков, то необходима поясняющая легенда;
- все линии на графиках должны быть чётко видны (нет похожих цветов или цветов, сливающихся с фоном);
- если отображена величина, имеющая очевидный диапазон значений (например, проценты могут быть от 0 до 100), то желательно масштабировать ось на весь диапазон значений (исключением является случай, когда вам необходимо показать малое отличие, которое незаметно в таких масштабах);
- сетка на графике помогает оценить значения в точках на глаз, это обычно полезно, поэтому лучше ее отрисовывать;
- если распределение на гистограмме имеет тяжёлые хвосты, лучше использовать логарифмическую шкалу.

Еще одна билиотека для визуализации: [seaborn](https://jakevdp.github.io/PythonDataScienceHandbook/04.14-visualization-with-seaborn.html) (обычно сокращается до sns). Это настройка над matplotlib, иногда удобнее и красивее делать визуализации через неё.

**5 пунктов после которых ваши графики не будут прежними:**
- для красивой картинки <code>%config InlineBackend.figure_format = 'retina'</code>
- задаем размер (почти) всех графиков (можно выбрать другие значения) <code>plt.rcParams['figure.figsize'] = 8, 5</code>
- размер шрифта подписей графиков <code>plt.rcParams['font.size'] = 12</code>
- формат в котором сохраняется изображение <code>mpl.rcParams['savefig.format'] = 'pdf'</code>

## 📌 **Алгоритм распределения студентов по курсам**:
1. По умолчанию на каждой дисциплине по выбору у 3 и 4 курсов может учиться 1 группа (до 30 студентов). Исключения описаны ниже. На blended-дисциплинах для четверокурсников количество мест не ограничено.
2. Проводится первая волна отбора. 3. На дисциплинах, где остались места после первой волны, формируются списки тех, кто выбрал их вторым приоритетом, и места заполняются лучшими по перцентили рейтинга студентами. После этого проводится такая же процедура для дисциплин третьего приоритета.
4. Если студент не попал на необходимое количество курсов по итогам трёх волн, с ним связывается учебный офис и решает вопрос в индивидуальном порядке.

#### 2. [2 балла] Распределите студентов в соответствии с первым приоритетом

```python
fall_course_capacity = {
    'Statistical Learning Theory': 60,
    'Высокопроизводительные вычисления': 60,
    'Анализ неструктурированных данных': float('inf'),
}

fall_courses = df[['fall_1', 'fall_2', 'fall_3']].stack().unique()
default_limit = 30
for course in fall_courses:
    if course not in fall_course_capacity:
        fall_course_capacity[course] = default_limit

fall_assigned_students = {course: [] for course in fall_course_capacity}

df = df.copy()
df.loc[:, 'course1'] = pd.NA
df.loc[:, 'course2'] = pd.NA
df.loc[:, 'is_first_place'] = pd.NA
df.loc[:, 'is_last_place'] = pd.NA

def first_wave_assignment(df, course_capacity, season_1, season_2, season_3, course_n1, course_n2, num_courses, assigned_students):
    for course in course_capacity:
        first_priority_students = df[(df[season_1] == course) |
                                     ((df[season_2] == course) & (df[num_courses] == 2))]

        first_priority_students = first_priority_students.sort_values('percentile')

        available_places = course_capacity[course]

        for idx, student in first_priority_students.iterrows():
            if available_places <= 0:
                break

            student_id = student['id']

            if student_id not in assigned_students[course]:
                if pd.isna(df.loc[idx, course_n1]):
                    df.loc[idx, course_n1] = course
                    assigned_students[course].append(student_id)
                    available_places -= 1
                elif pd.isna(df.loc[idx, course_n2]) and df.loc[idx, num_courses] == 2:
                    df.loc[idx, course_n2] = course
                    assigned_students[course].append(student_id)
                    available_places -= 1

        if not first_priority_students.empty:
            first_student_idx = assigned_students[course][0]
            df.loc[df['id'] == first_student_idx, 'is_first_place'] = True
            last_student_idx = assigned_students[course][-1]
            df.loc[df['id'] == last_student_idx, 'is_last_place'] = True


    return df

df = first_wave_assignment(df, fall_course_capacity, 'fall_1', 'fall_2', 'fall_3', 'course1', 'course2', 'num_fall_courses', fall_assigned_students)
```

#### 3. [2 балла] Проведите все три волны отбора студентов на курсы по выбору

```python
def second_wave_assignment(df, course_capacity, season_1, season_2, season_3, course_n1, course_n2, num_courses, assigned_students):

    for course in course_capacity:
        second_priority_students = df[((df[course_n1].isna()) & (df[season_2] == course) & (df[num_courses] == 1)) |
         ((df[season_3] == course) & (df[num_courses] == 2) & (df[course_n2].isna() | df[course_n1].isna()))]

        second_priority_students = second_priority_students.sort_values('percentile')

        available_places = course_capacity[course] - len(assigned_students[course])

        for idx, student in second_priority_students.iterrows():
            if available_places <= 0:
                break

            student_id = student['id']

            if student_id not in assigned_students[course]:
                if pd.isna(df.loc[idx, course_n1]):
                    df.loc[idx, course_n1] = course
                    assigned_students[course].append(student_id)
                    available_places -= 1
                elif pd.isna(df.loc[idx, course_n2]) and df.loc[idx, num_courses] == 2:
                    df.loc[idx, course_n2] = course
                    assigned_students[course].append(student_id)
                    available_places -= 1

    return df
```

```python
def third_wave_assignment(df, course_capacity, season_1, season_2, season_3, course_n1, course_n2, num_courses, assigned_students):

    for course in course_capacity:
        third_priority_students = df[((df[course_n1].isna()) & (df[season_3] == course) & (df[num_courses] == 1))]

        third_priority_students = third_priority_students.sort_values('percentile')

        available_places = course_capacity[course] - len(assigned_students[course])

        for idx, student in third_priority_students.iterrows():
            if available_places <= 0:
                break

            student_id = student['id']

            if student_id not in assigned_students[course]:
                if pd.isna(df.loc[idx, course_n1]):
                    df.loc[idx, course_n1] = course
                    assigned_students[course].append(student_id)
                    available_places -= 1

    return df
```
